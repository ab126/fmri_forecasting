```python
# CELL 1: IMPORT LIBRARIES

import os
import glob
import zipfile
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
# CELL 2: UNZIP THE DATASET

zip_path = "/content/pooled_stratified_share.zip"
extract_path = "/content/pooled_stratified_share"

# Check that the zip file exists before trying to unzip it
if not os.path.exists(zip_path):
    raise FileNotFoundError(
        f"Could not find zip file at {zip_path}. "
        "Please upload it into the Colab Files panel first."
    )

# Create the extraction folder if it does not already exist
os.makedirs(extract_path, exist_ok=True)

# Unzip the file
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)

print("Finished unzipping dataset.")
print("Extracted to:", extract_path)
```

```python
# CELL 3: FIND ALL .NPZ FILES

# Dataset is organized by this: subject/session/run.
# Each run is stored as one .npz file.
#
# This cell recursively searches the extracted folder and
# collects all .npz file paths.

npz_files = sorted(glob.glob(os.path.join(extract_path, "**", "*.npz"), recursive=True))

print(f"Found {len(npz_files)} .npz files.\n")
```

```python

# CELL 4: INSPECT ONE EXAMPLE FILE
# This cell loads one .npz file and prints:
# - the variable names inside it
# - shape and dtype of each array
# - a small preview when possible

example_file = npz_files[0]
example_data = np.load(example_file, allow_pickle=True)

print("Example file:")
print(example_file)
print("\nVariables inside:")
print(example_data.files)
print()

for key in example_data.files:
    value = example_data[key]
    print(f"Variable: {key}")
    print(f"  Type:  {type(value)}")

    if isinstance(value, np.ndarray):
        print(f"  Shape: {value.shape}")
        print(f"  Dtype: {value.dtype}")

        # Show a small preview to help understand the content
        if value.size <= 20:
            print(f"  Values: {value}")
        else:
            print(f"  First few values: {value.flatten()[:10]}")
    else:
        print(f"  Value: {value}")

    print("-" * 50)
```

```python

# CELL 5: PLOT BOLD SIGNALS FOR ONE RUN

timeseries = example_data["timeseries"]

print("timeseries shape:", timeseries.shape)

plt.figure(figsize=(12, 6))

for i in range(timeseries.shape[0]):
    plt.plot(timeseries[i], alpha=0.7, label=f"ROI {i}" if i < 5 else None)

plt.title("BOLD signals across ROIs for one run")
plt.xlabel("Timepoint")
plt.ylabel("Signal")
plt.show()
```

The input data consists of pre-parcellated fMRI time series with shape (number of ROIs × timepoints). Each run is first standardized using per-ROI z-scoring. The data is then parsed into supervised learning samples using a sliding window approach, where a sequence of past timepoints across all ROIs is used to predict the next timepoint. These samples are aggregated across runs, and a run-level split is used to ensure that training and testing are performed on separate runs, preventing data leakage.

```python
# CELL 7: DEFINE A FUNCTION TO BUILD FORECASTING SAMPLES

#
# Original GitHub idea:
#   stimulus features -> future BOLD
#
# Our lab's data:
#   past BOLD -> future BOLD
#
# So we convert one run of shape:
#   (num_rois, num_timepoints) into many supervised learning samples using a sliding window.
# Since each timepoint is a vector of all ROIs, each input is
# a window of brain states, and each target is the future brain state. (same thing as the transformer cell 5)

def make_windows_from_timeseries(timeseries, window_size=5, horizon=1):
    """
    Convert one run's ROI time series into supervised forecasting samples.

    Parameters
    ----------
    timeseries : np.ndarray
        Shape: (num_rois, num_timepoints)
    window_size : int
        Number of past timepoints used as input
    horizon : int
        How many steps into the future to predict

    Returns
    -------
    X : np.ndarray
        Shape: (num_samples, window_size * num_rois)
        We flatten the input window so it can be used by a linear model.
    Y : np.ndarray
        Shape: (num_samples, num_rois)
        Future ROI vector we want to predict.
    """

    num_rois, num_timepoints = timeseries.shape

    X_list = []
    Y_list = []

    # We slide a window through time.
    # The last valid starting point is limited by both the window size
    # and the prediction horizon.
    for start_t in range(num_timepoints - window_size - horizon + 1):
    # samples = timepoints − window_size − horizon + 1
        end_t = start_t + window_size
        target_t = end_t + horizon - 1

        # Input window shape before flattening:
        x_window = timeseries[:, start_t:end_t]
        y_target = timeseries[:, target_t]
        x_flat = x_window.flatten()

        X_list.append(x_flat)
        Y_list.append(y_target)

    X = np.array(X_list)
    Y = np.array(Y_list)

    return X, Y


    # Use the timeseries loaded earlier from the npz file
X, Y = make_windows_from_timeseries(timeseries, window_size=5, horizon=1)

print("Input shape (X):", X.shape)
print("Target shape (Y):", Y.shape)
```

```python

# CELL 8: DETERMINE ROI COUNT CONSISTENCY

# Some runs contain different numbers of ROIs (e.g., 18 vs 19).
# To make sure all runs have identical feature dimensions,
# we detect the smallest ROI count across the dataset.

# Similar to transformer Cell 12


roi_counts = []

for file_path in npz_files:

    data = np.load(file_path, allow_pickle=True)
    timeseries = data["timeseries"]

    roi_counts.append(timeseries.shape[0])

roi_counts = np.array(roi_counts)

print("ROI counts found in dataset:", np.unique(roi_counts))

min_rois = np.min(roi_counts)

print("\nMinimum ROI count used for modeling:", min_rois)
```

```python

# CELL 9: BUILD FORECASTING DATASET

# Steps:
# 1. Load each run
# 2. Truncate ROIs to min_rois
# 3. Build sliding windows
# 4. Store samples

window_size = 30 # Changed to 30 to fit the regression model
horizon = 1

X_all = []
Y_all = []
run_ids = []

for run_idx, file_path in enumerate(npz_files):

    data = np.load(file_path, allow_pickle=True)
    timeseries = data["timeseries"]


    timeseries = timeseries[:min_rois]
    # z-score each ROI across time so preprocessing matches transformer notebook better
    mean = timeseries.mean(axis=1, keepdims=True)
    std = timeseries.std(axis=1, keepdims=True)
    timeseries = (timeseries - mean) / (std + 1e-8)


    X_run, Y_run = make_windows_from_timeseries(
        timeseries,
        window_size=window_size,
        horizon=horizon
    )

    X_all.append(X_run)
    Y_all.append(Y_run)

    run_ids.extend([run_idx] * len(X_run))

# Stack all runs together
X_all = np.vstack(X_all)
Y_all = np.vstack(Y_all)
run_ids = np.array(run_ids)

print("Dataset successfully built")
print("X_all shape:", X_all.shape)
print("Y_all shape:", Y_all.shape)
print("Total runs used:", len(np.unique(run_ids)))

# Result:
# X_all : (num_samples, window_size * min_rois)
# Y_all : (num_samples, min_rois)
```

```python
# CELL 10: RUN-SAFE TRAIN TEST SPLIT

unique_runs = np.unique(run_ids)

# Shuffle runs
np.random.seed(42)
np.random.shuffle(unique_runs)

split_idx = int(0.8 * len(unique_runs))

train_runs = unique_runs[:split_idx]
test_runs = unique_runs[split_idx:]

train_mask = np.isin(run_ids, train_runs)
test_mask = np.isin(run_ids, test_runs)

X_train = X_all[train_mask]
Y_train = Y_all[train_mask]

X_test = X_all[test_mask]
Y_test = Y_all[test_mask]

print("Train samples:", X_train.shape[0])
print("Test samples:", X_test.shape[0])
```

**We split the dataset at the run level using a fixed random seed (42) and an 80/20 split. This means all samples from a given run are either entirely in the training set or entirely in the test set, preventing data leakage across runs. The same split is used consistently across models for fair comparison. For evaluation, we construct sliding window samples from the test runs and compute MSE on all test windows, so performance reflects how well the model generalizes to unseen runs.**

```python
# CELL 11: FEATURE SCALING

# Standardize input features using training data statistics.
# This improves stability of regression models.


from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print("Feature scaling completed")
```

```python
# CELL 12: TRAIN LINEAR RIDGE MODEL

# simple ridge regression

from sklearn.linear_model import Ridge

alpha = 1.0

model = Ridge(alpha=alpha)
model.fit(X_train_scaled, Y_train)

print("Model training completed")
```

```python
# CELL 13: MODEL EVALUATION

# Predict the test data and compute mean squared error.

from sklearn.metrics import mean_squared_error

Y_pred = model.predict(X_test_scaled)

mse = mean_squared_error(Y_test, Y_pred)

print("Test MSE:", mse)

# per-ROI error
roi_mse = np.mean((Y_test - Y_pred) ** 2, axis=0)

print("\nPer-ROI MSE:")
for i, val in enumerate(roi_mse):
    print(f"ROI {i}: {val:.6f}")
```

```python
# CELL 14: BASELINE FORECAST

num_rois = min_rois

X_test_reshaped = X_test.reshape(-1, num_rois, window_size)
baseline_pred = X_test_reshaped[:, :, -1]

baseline_mse = mean_squared_error(Y_test, baseline_pred)

print("Baseline MSE:", baseline_mse)
print("Linear model MSE:", mse)
```

```python
# CELL 15: VISUALIZE PREDICTIONS

# Plot true vs predicted signals for several ROIs.


plt.figure(figsize=(12,8))

num_rois_to_plot = min(4, Y_test.shape[1])

for i in range(num_rois_to_plot):

    plt.subplot(num_rois_to_plot,1,i+1)

    plt.plot(Y_test[:100, i], label="True")
    plt.plot(Y_pred[:100, i], label="Predicted")

    plt.title(f"ROI {i}")
    plt.legend()

plt.tight_layout()
plt.show()
```

# **Statistics**

```python
# CELL 21: ORGANIZE TEST PREDICTIONS AND RUN LABELS
# This cell gathers the true test values, predicted test values,
# and the run ID for each test sample.
# We will use these arrays in the next cells for overall, ROI-wise,
# and run-level statistics.

X_train = X_all[train_mask]
Y_train = Y_all[train_mask]

X_test = X_all[test_mask]
Y_test = Y_all[test_mask]

print("X_test shape:", X_test.shape)

print("Y_test shape:", Y_test.shape)
Y_test_all = Y_test.copy()
Y_pred_all = Y_pred.copy()

# run_ids contains the run label for every sample in X_all / Y_all
# test_mask selects only the samples that belong to held-out test runs
test_sample_run_ids = run_ids[test_mask]

print("Y_test_all shape:", Y_test_all.shape)
print("Y_pred_all shape:", Y_pred_all.shape)
print("test_sample_run_ids shape:", test_sample_run_ids.shape)

print("Number of held-out test runs:", len(np.unique(test_sample_run_ids)))
```

```python
# CELL 22: OVERALL TEST MSE AND SCALED MSE
# This cell computes:
# 1. Overall test MSE across all ROIs and all test samples
# 2. A scaled MSE using a naive one-step difference scale from the training runs
#
# The scaled MSE is useful because it tells us how large the model error is
# relative to the natural step-to-step variability of the training signal.

from sklearn.metrics import mean_squared_error
import numpy as np

# Overall test MSE
overall_mse = mean_squared_error(Y_test_all, Y_pred_all)

# Compute scaling denominator from TRAIN runs only
# We use the average squared first difference across all ROIs in all training runs.
train_naive_sq_diffs = []

for run_idx in train_runs:
    data = np.load(npz_files[run_idx], allow_pickle=True)
    timeseries = data["timeseries"][:min_rois]

    # same preprocessing as the forecasting dataset
    mean = timeseries.mean(axis=1, keepdims=True)
    std = timeseries.std(axis=1, keepdims=True)
    timeseries = (timeseries - mean) / (std + 1e-8)

    diffs = np.diff(timeseries, axis=1)   # shape: (num_rois, time-1)
    train_naive_sq_diffs.append(diffs**2)

train_naive_sq_diffs = np.concatenate(
    [arr.reshape(-1) for arr in train_naive_sq_diffs]
)

scale_denom = np.mean(train_naive_sq_diffs)
scaled_mse = overall_mse / scale_denom

print("Overall Test MSE:", overall_mse)
print("Scale Denominator (train naive diff^2 mean):", scale_denom)
print("Scaled MSE:", scaled_mse)
```

```python
# CELL 23: PER-ROI PERFORMANCE STATISTICS
# This cell computes test MSE separately for each ROI.
# This helps us see whether the model performs consistently across brain regions
# or if some ROIs are much easier/harder to predict than others.

roi_mse = np.mean((Y_test_all - Y_pred_all) ** 2, axis=0)

print("Per-ROI MSE:")
for roi_idx, val in enumerate(roi_mse):
    print(f"ROI {roi_idx}: {val:.6f}")

print("\nSummary across ROIs:")
print("Mean ROI MSE:", np.mean(roi_mse))
print("Std ROI MSE:", np.std(roi_mse))
print("Best ROI MSE:", np.min(roi_mse))
print("Worst ROI MSE:", np.max(roi_mse))
```

```python
# CELL 24: RUN-LEVEL PERFORMANCE ACROSS HELD-OUT TEST RUNS
# This cell computes one MSE value per held-out test run.
# Since each run contributes many sliding-window test samples,
# we group those samples back by run ID and measure how well the model
# performs on each unseen run individually.

unique_test_runs = np.unique(test_sample_run_ids)

run_level_mse = []
run_level_num_samples = []

for run_id in unique_test_runs:
    run_mask = (test_sample_run_ids == run_id)

    y_true_run = Y_test_all[run_mask]
    y_pred_run = Y_pred_all[run_mask]

    mse_run = mean_squared_error(y_true_run, y_pred_run)

    run_level_mse.append(mse_run)
    run_level_num_samples.append(np.sum(run_mask))

run_level_mse = np.array(run_level_mse)
run_level_num_samples = np.array(run_level_num_samples)

print("Run-level MSE for each held-out test run:")
for i, run_id in enumerate(unique_test_runs):
    print(f"Run {run_id}: MSE = {run_level_mse[i]:.6f} | Samples = {run_level_num_samples[i]}")

print("\nRun-level summary:")
print("Mean run MSE:", np.mean(run_level_mse))
print("Std run MSE:", np.std(run_level_mse))
print("Best run MSE:", np.min(run_level_mse))
print("Worst run MSE:", np.max(run_level_mse))
```

```python
# CELL 25: PLOT PER-ROI MSE
# This plot shows how prediction error changes across ROIs.

import matplotlib.pyplot as plt

plt.figure(figsize=(10, 4))
plt.bar(np.arange(len(roi_mse)), roi_mse)
plt.xlabel("ROI index")
plt.ylabel("Test MSE")
plt.title("Linear Regression: Per-ROI Test MSE")
plt.show()
```

```python
# CELL 26: PLOT RUN-LEVEL MSE
# This plot shows how prediction performance varies from one held-out run to another.

plt.figure(figsize=(10, 4))
plt.plot(run_level_mse, marker='o')
plt.xlabel("Held-out test run index")
plt.ylabel("Run-level MSE")
plt.title("Linear Regression: Run-Level Test MSE Across Held-Out Runs")
plt.show()
```

```python
# CELL 27: SAVE LINEAR REGRESSION STATISTICS FOR LATER MODEL COMPARISON
# This cell saves the key outputs and summary statistics so they can later be loaded
# into a separate comparison notebook alongside the transformer and exponential smoothing results.

linear_results = {
    "model_name": "Linear Regression (Ridge)",
    "window_size": window_size,
    "horizon": horizon,
    "num_rois": min_rois,

    "overall_mse": overall_mse,
    "scaled_mse": scaled_mse,

    "roi_mse": roi_mse,
    "mean_roi_mse": np.mean(roi_mse),
    "std_roi_mse": np.std(roi_mse),

    "run_level_mse": run_level_mse,
    "mean_run_mse": np.mean(run_level_mse),
    "std_run_mse": np.std(run_level_mse),

    "Y_test_all": Y_test_all,
    "Y_pred_all": Y_pred_all,
    "test_sample_run_ids": test_sample_run_ids,
    "test_runs": unique_test_runs
}

np.save("/content/linear_regression_statistics.npy", linear_results, allow_pickle=True)

print("Saved to: /content/linear_regression_statistics.npy")
```

These statistics summarize linear regression performance at multiple levels. Overall MSE and scaled MSE give a global measure of forecasting accuracy, per-ROI MSE shows whether some brain regions are easier or harder to predict, and run-level MSE shows how stable the model is across different held-out runs. Saving these outputs now makes it easier to compare this model fairly with the transformer and exponential smoothing models later in one separate notebook.
